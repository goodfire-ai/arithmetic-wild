Code snippet to load the best DAS subspaces that we trained, stored on Hugging Face Hub.

See Appendix D.3 for details about these subspaces.

```python
import yaml
from utils import load_subspace_hf

# always constant for this project 
model_name = "Llama-3.1-8B"
LAYER = 18

# choose to load in output post-MLP subspaces.
subspace_type = "output" 
# subspace_type = "input" # uncomment to load `input concept` subspace pre-MLP 
# subspace_type = "offset" # uncomment to load `offset` subspace pre-MLP 
```

```python
subspaces = {}
for task in ["months", "weekdays", "hours", "addition"]:
    print(task, subspace_type)
    config_path = f"../best_das/{model_name}/L{LAYER}_{task}.yaml"
    with open(config_path) as f:
        config = yaml.safe_load(f)

    sub_cfg = config["subspaces"][subspace_type]
    subspace = load_subspace_hf(task, subspace_type, layer=LAYER)
    subspaces[task] = subspace 
    print(f"Loaded {task} {subspace_type} subspace: {subspace.shape}")
    print(f"  IIA: {sub_cfg['subspace_iia']:.3f}")
```
